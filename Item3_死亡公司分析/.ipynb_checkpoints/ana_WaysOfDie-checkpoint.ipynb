```python
import pandas as pd
import numpy as np
from pyecharts import WordCloud
```

```python
df = pd.read_csv("company_data.csv",encoding="utf-8")
t = df.iloc[:,1] # 第二个字段：死亡原因
t1 = t.copy()
t1
```

```python
all_list = []
for i in t1:
    one_words = i.split("/")
    all_list += one_words
print(all_list)

```

```python
# 创建DataFrame对象，并统计词的次数
# word作为列名
res = pd.DataFrame(all_list,columns=["word"])
# 对每个词进行分组，并统计其出现的次数
res = res["word"].groupby(res["word"]).count()

res.index.name = "txt"
res = res.reset_index()

name = res["txt"].tolist()
number = res["word"].tolist()

wc = WordCloud(width=1500,height=1000)
wc.add("",name,number,word_size_range=[10,100])
wc.render("死亡公司原因分析.html")
```
