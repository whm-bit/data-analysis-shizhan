上映后每天的评论数量走势

上映后每天的评分走势

评分饼图

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nezha.csv",encoding="utf-8",names=["date","rate","title","users","name","comment"],usecols=[0,1,2,3,4,5])
t1 = df.copy()
t1
```

```python
t1.dropna(axis=0,inplace=True) #删除缺失值，0指行，1指列 True表示就地更改
t1
```

```python
# 处理日期字段，保留年月日，去掉时分秒
t1["date"] = t1["date"].apply(lambda x: str(x).split(" ")[0])
t1["count"] =1
t1
```

```python
# 筛选出上映之后的评论数据
t2 = t1[t1["date"] > "2019-07-26"]
t2
```

下面是具体需求

```python
# 上映后每天的评论数量走势
# 按天统计评论的个数
t2_result = t2.groupby("date")["count"].agg(["count"]) #聚合函数,后面的count是统计数量
# t2_result
t2_result.plot(kind='bar')
plt.show()
```

```python
# 上映后每天的评分走势
# 统计每天评论的平均分
t3_result = t2.groupby("date")["rate"].agg(["mean"])
t3_result.plot(kind='bar')
plt.show()
```

```python
# 评分饼图
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号‘-’显示为方块的问题

# 统计评论的个数
t4_result = t2.groupby("rate")["count"].agg(["count"])

rates =t4_result.index.tolist() # 分数转化成列表
numbers = [one_list[0] for one_list in t4_result.values.tolist()] # 每个分数对应的数量,也转化成一个列表
# numbers
# rates
explode = (0,0,0.1,0,0)
plt.pie(numbers,labels=rates,explode=explode,autopct='%1.2f')
plt.title("评分饼图",color='white')

# plt.axis('equal') # 上图是一个椭圆，加上这个条件就是正圆形
plt.show()

"""
pie函数格式
 def pie
 (x, explode=None, labels=None, colors=None, autopct=None,
 pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None,
 radius=None, counterclock=True, wedgeprops=None, textprops=None,
 center=(0, 0), frame=False, rotatelabels=False, hold=None, data=None)



pie 函数参数详解
 x       :(每一块)的比例，如果sum(x) > 1会使用sum(x)归一化；(具体的结果数值)
 labels  :(每一块)饼图外侧显示的说明文字；(每一个数值的标签)
 explode :(每一块)离开中心距离；
 startangle :起始绘制角度,默认图是从x轴正方向逆时针画起,如设定=90则从y轴正方向画起；
 shadow  :在饼图下面画一个阴影。默认值：False，即不画阴影；
 labeldistance :label标记的绘制位置,相对于半径的比例，默认值为1.1, 如<1则绘制在饼图内侧；
 autopct :控制饼图内百分比设置,可以使用format字符串或者format function
         '%1.1f'指小数点前后位数(没有用空格补齐)；
 pctdistance :类似于labeldistance,指定autopct的位置刻度,默认值为0.6；
 radius  :控制饼图半径，默认值为1；
 counterclock ：指定指针方向；布尔值，可选参数，默认为：True，即逆时针。将值改为False即可改为顺时针。
 wedgeprops ：字典类型，可选参数，默认值：None。参数字典传递给wedge对象用来画一个饼图。例如：wedgeprops={'linewidth':3}设置wedge线宽为3。
 textprops ：设置标签（labels）和比例文字的格式；字典类型，可选参数，默认值为：None。传递给text对象的字典参数。
 center ：浮点类型的列表，可选参数，默认值：(0,0)。图标中心位置。
 frame ：布尔类型，可选参数，默认值：False。如果是true，绘制带有表的轴框架。
 rotatelabels ：布尔类型，可选参数，默认为：False。如果为True，旋转每个label到指定的角度。

"""
```
