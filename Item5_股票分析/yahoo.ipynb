```python
# 基本信息
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

# 股票数据的获取
import pandas_datareader as pdr

# 可视化
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# time
from datetime import datetime

```

```python
start = datetime(2016,9,20)
alibaba = pdr.get_data_yahoo('BABA',start=start)
amazon = pdr.get_data_yahoo('AMZN',start=start)
print("over")
```

```python
alibaba.to_csv('BABA.csv')
amazon.to_csv('AMZN.csv')
print("OVER ")
```

```python
# 查看股价走势图
alibaba['Adj Close'].plot(legend=True)

```

```python
alibaba['Adj Close'].plot()
amazon['Adj Close'].plot()
```

```python
alibaba['high-low'] = alibaba['High'] - alibaba['Low']
```

```python
alibaba.head()
```

```python
alibaba['high-low'].plot()
```

```python
# daily return 天与天之间的变化
alibaba['daily-return'] = alibaba['Adj Close'].pct_change()
```

```python
alibaba.head()
```

```python
alibaba['daily-return'].plot()
# 结果有正有负，说明天与天之间有涨有跌
```

```python
alibaba['daily-return'].plot(figsize=(10,4)) # 将图形变长
```

```python
alibaba['daily-return'].plot(figsize=(10,4),linestyle='--',marker='o') # 将线条变成--，将高低转折点端点着重用‘o’表示
```

```python
alibaba['daily-return'].plot(kind='hist') # 直方图
```

```python
sns.distplot(alibaba['daily-return'].dropna(),bins=100,color='purple') # 删除缺失值
```
