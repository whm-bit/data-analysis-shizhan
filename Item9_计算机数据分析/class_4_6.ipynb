```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_excel('计算机与信息技术学院四六级情况.xlsx',header=None)
df1
```

```python
df1 = df1.drop(index=[0]) # 删除第一行
```

```python
df1 = df1.rename(columns={0:"班级",1:"学号",2:"姓名",3:"项目",4:"四级",5:"六级"})
```

```python

df1 = df1.drop(index=[1]) # 删除行名是1的那一行
df1
```

```python
df2 = df1[["班级","学号","姓名","四级","六级","项目"]]
```

```python
df2.index = [i_index for i_index in range(0,89)]
```

```python
df2 # 数据整理完毕
```

```python
def banji(bj):
    if bj == "17计科一班":
        return pd.Series([1])
    elif bj == "17计科二班":
        return pd.Series([2])
    elif bj == "17计科三班":
        return pd.Series([3])
```

```python
df2["班级"] = df2["班级"].apply(banji) # 将班级列里面的文字转换为数字
```

```python
df2
```

```python
def siliu(sl):
    if sl=="是":
        return pd.Series([1])
    elif sl == "否":
        return pd.Series([0])
```

```python
df2["四级"] = df2["四级"].apply(siliu)
df2["六级"] = df2["六级"].apply(siliu)
```

```python
df2 # 将四六级列里面的是和否转化成1和0
```

```python
def xiangmu(xm):
    if xm == "是":
        return pd.Series([1])
    elif xm == "无":
        return pd.Series([0])
```

```python
df2["项目"] = df2["项目"].apply(xiangmu)
```

```python
df2 # 整理完毕
```

```python
df2.to_csv('Class_4_6.csv')
```
